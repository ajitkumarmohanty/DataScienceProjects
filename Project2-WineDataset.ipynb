```python
# Checking the skewness of the dataset
winedf.skew()

# so here there are few variable where the skewness is more than 0.55, so we have to treat them.
# the variables are Malic acid, Magnesium, Color intensity  and Proline
```

```python
# treating the skewness using log
import numpy as np
for col in winedf.columns:
    if winedf.skew().loc[col]>0.55:
        winedf[col]=np.log1p(winedf[col])
```

```python
#checking the outliers
from scipy.stats import zscore
print(winedf.shape)
z_score=abs(zscore(winedf))
print('Z Score is:','\n',z_score)
winedf_final=winedf.loc[(z_score<3).all(axis=1)]
print(winedf_final.shape)
```

```python
# separating the input and output/target variables
import numpy as np
df_x=winedf_final.drop(columns=['Class'])
y=winedf_final[['Class']]

print(df_x.shape)
print(y.shape)
```

```python
#scaling the input variables
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
```

```python
# Again checking the skewness of the dataset
x.skew()
```

```python
#dividing the train and test dataset and find out the r2 score corresponding to random state

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
max_r2_score=0.0
for r_state in range(42,101):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=r_state)
    regr=linear_model.LinearRegression()
    regr.fit(x_train,y_train)
    y_pred=regr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print('R2 score corresponding to random state:',r_state, 'is:',r2_scr)
    if r2_scr>max_r2_score:
        max_r2_score=r2_scr
        final_r_state=r_state
print('Maximum R2 score corresponding to random state',final_r_state,'is:',max_r2_score)
```

```python
#checking the overfitting and underfitting using cross validation
from sklearn.model_selection import cross_val_score
cross_val_score(linear_model.LinearRegression(),x,y,cv=5,scoring='r2').mean()
```

```python
# Making the Final Model at random_state=43
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=43)
lr=linear_model.LinearRegression()
lr.fit(x_train,y_train)
y_pred=regr.predict(x_test)
```

```python
#Evaluating the Final Model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
print('r2 score is:',r2_score(y_test,y_pred))
print('RMSE is:',np.sqrt(mean_squared_error(y_test,y_pred)))
```
