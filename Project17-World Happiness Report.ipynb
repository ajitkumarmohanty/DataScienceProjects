```python
#So here for our analysis we can drop Country value
happiness=happiness.drop('Country',axis=1)
```

```python
#here there are more than one Region value, so lets encode this value for our analysis
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
happiness['Region']=le.fit_transform(happiness['Region'])
happiness.head(10)
```

```python
#here Happiness Rank value is not required for our analysis, so lets drop this column also
happiness=happiness.drop('Happiness Rank',axis=1)
happiness.head(5)
```

```python
#lets print the Happiness Score descrption
print(happiness['Happiness Score'].describe())
```

```python
#checking the outliers
from scipy.stats import zscore
print(happiness.shape)
z_score=abs(zscore(happiness))
print('Z Score is:','\n',z_score)
happiness_final=happiness.loc[(z_score<3).all(axis=1)]
print(happiness_final.shape)
```

```python
# Checking the skewness of the dataset
happiness_final.skew()
```

```python
# so here there are few variable where the skewness is more than 0.55, so we have to treat them.
# treating the skewness using log
import numpy as np
for col in happiness_final.columns:
    if happiness_final.skew().loc[col]>0.55:
        happiness_final[col]=np.log1p(happiness_final[col])
```

```python
# Again checking the skewness of the dataset
happiness_final.skew()
```

```python
# separating the input and output/target variables
import numpy as np
dx=happiness_final.drop(columns=['Happiness Score'],axis=1)
y=happiness_final[['Happiness Score']]

print(dx.shape)
print(y.shape)
```

```python
#scaling the input variables
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(dx)
x=pd.DataFrame(x,columns=dx.columns)
```

```python
#dividing the train and test dataset
from sklearn.model_selection import train_test_split, cross_val_score
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
#dividing the train and test dataset and find out the r2 score corresponding to random state

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
def max_r2_score(regr,x,y):
    max_r_score=0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=r_state)
        regr.fit(x_train,y_train)
        y_pred=regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        #print('R2 score corresponding to random state:',r_state, 'is:',r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print('Maximum R2 score corresponding to random state',final_r_state,'is:',max_r_score)
```

```python
#lets use the linear Regression model
from sklearn.linear_model import LinearRegression
lreg=LinearRegression()
r_state=max_r2_score(lreg,x,y)
```

```python
#lets use grid search method to find optimal value of n_neighbors for knn model
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
neighbors={'n_neighbors': range(1,30)}
knn=KNeighborsRegressor()
gknr=GridSearchCV(knn,neighbors,cv=10)
gknr.fit(x,y)
gknr.best_params_
```

```python
#lets build the model using 'n_neighbors': 10
knn=KNeighborsRegressor(n_neighbors=10)
r_state=max_r2_score(knn,x,y)
```

```python
#checking the mean r2 score and overfitting and underfitting using cross validation
from sklearn.model_selection import cross_val_score
print('Mean R2 score of Linear Regression is:', cross_val_score(lreg,x,y,cv=5,scoring='r2'.mean()))
print('Standard Deviation of Linear Regression is:',cross_val_score(lreg,x,y,cv=5,scoring='r2'.std()))
```

```python
# lets put this alpha value in the lasso model and build the model
lsreg=Lasso(alpha=0.001)
r_state=max_r2_score(lsreg,x,y)
```

```python
# lets put this alpha value in the Ridge model and build the model
lrreg=Ridge(alpha=0.001)
r_state=max_r2_score(lrreg,x,y)
```

```python
#defining the models which will be used for model building
from sklearn.linear_model import LinearRegression
lreg=LinearRegression()
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
from sklearn.linear_model import Lasso
lsreg=Lasso(alpha=0.1)
from sklearn.linear_model import Ridge
lrreg=Ridge(alpha=0.1)
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor()
from sklearn.ensemble import AdaBoostRegressor
ada=AdaBoostRegressor()
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
```

```python
#appending all the models
models=[]
models.append(('LinearRegression',lreg))
models.append(('KNeighborsRegressor',knn))
models.append(('Lasso',lsreg))
models.append(('Ridge',lrreg))
models.append(('GradientBoostingRegressor',gbr))
models.append(('AdaBoostRegressor',ada))
models.append(('RandomForestRegressor',rf))
```

```python
#printing the output of each model
Model=[]
RState=[]
score=[]
for name, model in models:
    print('************************',name,'**********************')
    print('\n')
    Model.append(name)
    max_r2_score=0.0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=r_state)
        model.fit(x_train,y_train)
        y_pred=model.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        if r2_scr>max_r2_score:
            max_r2_score=r2_scr
            final_r_state=r_state
    print('Final Random State is:', final_r_state)
    print('\n')
    RState.append(final_r_state*1)
    print('Maximum R2 Score is:', max_r2_score)
    print('\n')
    score.append(max_r2_score*100)
    print('\n')
```

```python
#lets print the final result of all the models
result=pd.DataFrame({'Model':Model,'Random_State':RState,'R2_Score':score})
print(result)
```
